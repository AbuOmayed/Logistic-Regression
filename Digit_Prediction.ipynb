```python
#import package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
#Import dataset
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
digits
```

```python
digits.data.shape
```

```python
digits.target.shape
```

```python
len(digits.feature_names)
```

```python
#displaying some of the images and labels

plt.figure(figsize = (20,4))
for index, (image, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):

    plt.subplot(1,5, index+1)
    plt.imshow(np.reshape(image, (8,8)), cmap = plt.cm.gray)
    plt.title("Training %i\n"%label, fontsize = 20)
```

```python
#Devided dataset into train and test

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, random_state = 2, test_size = .30)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#Import the Logistic Regression model
from sklearn.linear_model import LogisticRegression
```

```python
#Making an instance of the model and training it
logreg = LogisticRegression(solver='liblinear')  # 'liblinear' solver:
                                                       #which is often faster and more robust for large-scale problems
logreg.fit(x_train, y_train)
```

```python
#predict the ouput of the first element of the test set
print(logreg.predict(x_test[0].reshape(1, -1)))
```

```python
#Predict the entire dataset
predict = logreg.predict(x_test)
```

```python
predict
```

```python
#Determine the accuary of the train data
print(logreg.score(x_train, y_train))
```

```python
#Determine the accuracy of the model
print(logreg.score(x_test, y_test))
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, predict)
```

```python
cm
```

```python
#Visualization confusion matrix using heatmap

plt.figure(figsize = (8,8))
sns.heatmap(cm, annot = True , fmt = ".3f", linewidth = .5, square = True, cmap = "Blues_r")
plt.xlabel("Actual label")
plt.ylabel("Predicted label")
plt.show()
```

```python
#Visually see how accurately fit the model

index = 0
classifiedindex = []
for pre, actual in zip(predict, y_test):
    if pre == actual:
        classifiedindex.append(index)
    index += 1   
plt.figure(figsize = (20,3))
for i , wrong in enumerate(classifiedindex[0:4]):
    plt.subplot(1,4, i+1)
    plt.imshow(np.reshape(x_test[wrong], (8,8)), cmap = plt.cm.gray)
    plt.title("Predict : {}, Actual : {}".format(predict[wrong], y_test[wrong]), fontsize = 20)
plt.show()
```
